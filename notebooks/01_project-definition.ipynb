## ***_info________________________________________***

### graph notation conventions
#### *standard flowchart notation (ISO 5807)*
- **Rounded rectangles** for nodes (agent steps / processing)
- **Diamonds** for conditional routing (yes â€” you're right on that)
- **Arrows** for edges, labeled with the condition or trigger
- **Stadium/pill shapes** for start and end points
- **Rectangles** for I/O (the brief in, concept out)
- Additional
    - **human-in-the-loop interrupt points indicator** - Use a distinct visual indicator for (e.g. a small pause icon or a different border style)
    - **feedback loops** - Use a max iteration count or an exit condition

#### *tooling*
- `FigJam` / `Miro`
    - Fine for sketch
- `Mermaid`
    - Markdown-compatible diagramming syntax that renders natively on GitHub. 
    - Architecture diagram lives as code in your docs, stays version-controlled, and updates with your project.

#### *general colour coding guidance*
- Note:
    - No strict rules.
    - Only encode *one* dimension of meaning.
    - Diagram should still read clearly in greyscale.
- Functional approach:
    - **green** - start/end points (input/output)
    - **blue** - processing nodes (agents)
    - **amber** / **yellow** - decision points
    - **red** / **orange** - error/fallback paths
- Mermaid style application example:
    ```
        graph TD
            A[Client Brief]:::input --> B[Strategist]:::agent
            B --> C[Creative]:::agent
            C --> D{Creative Director Review}:::decision
            D -->|Approved| E[Creative Concept]:::output
            D -->|Revise| C

            classDef input fill:#2d6a4f,color:#fff
            classDef agent fill:#1d3557,color:#fff
            classDef decision fill:#e9c46a,color:#000
            classDef output fill:#2d6a4f,color:#fff
    ```

## ***_data model________________________________________***

### flow
- **Input** -> the system receives a client brief (raw text)
- **Strategist** -> reads the client brief, produces a creative brief
- **Creative** -> reads the creative brief, generates ideas/concepts
- **Creative Director** -> reads the brief + the creative output, produces a score and feedback
- **Decision** -> checks the score and iteration count, routes accordingly
- **Output** -> the final approved creative concept

### state model requirements
- The original client brief
- The creative brief (strategist output)
- The current creative concept (creative output)
- The CD's evaluation (score + feedback) [structured]
- History of creative iterations
- An iteration counter
- The status of the overall workflow
- LLM used tracker
